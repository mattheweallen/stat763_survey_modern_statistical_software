##### Setup

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
iris = sns.load_dataset("iris")
```

##### The bar chart of counts of each species in the iris data set show equal number of samples.

```python
sns.countplot(x="species", data=iris)
plt.show()
```

#### From all the pair wise distributions, there appears to be a good separation of iris species by their relationship betwen sepal length and petal length

```python
sns.pairplot(iris, hue="species")
plt.show()
```

#### The scatter plot below shows good separation of iris species by their relationship betwen sepal length and petal length.

```python
sns.lmplot(data=iris,
           y='petal_length',
           x='sepal_length',
           hue='species')
plt.ylabel('Petal Length (cm)')
plt.xlabel('Sepal Length (cm)')
plt.title("Iris Petal Length vs Sepal Length by Species")
plt.show()
```
